# Generating Data

```python
import numpy as np

# Generate random noisy data
np.random.seed(0)
x = np.linspace(0, 18, 1000)
y = []
labels = []
for i in range(4):
    labels.append(f"Data {i}")

y.append(np.random.normal(0, 0.1, 1000))
y.append(np.random.normal(0, 0.1, 1000))
y.append(np.random.normal(0, 0.1, 1000))
y.append(np.random.normal(0, 0.1, 1000))
```

# Matplotlib

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(
    figsize=(27, 10)
)

for i, label in enumerate(labels):
    ax.plot(x, y[i] + i * 0.5, label=label)

ax.set_xlabel("Size (m^2)")
ax.set_ylabel("Efficiency (%)")
ax.legend()
plt.show()
```

# Manim

```python
from IPython.display import HTML
from manim import *
```

```python
# Function to render a scene in Jupyter notebook
def render_scene(scene_class):
    scene = scene_class()
    scene.render()
    return HTML(scene.renderer.file_writer.html_video())
```

```python


class BraceAnnotation(Scene):
    def construct(self):
        # Create dots
        dot = Dot([-2, -1, 0])
        dot2 = Dot([2, 1, 0])

        # Create line between dots
        line = Line(dot.get_center(), dot2.get_center()).set_color(ORANGE)

        # Create horizontal brace with text
        b1 = Brace(line, direction=DOWN)
        b1text = b1.get_text("Horizontal distance")

        # Create vertical brace with text
        b2 = Brace(line, direction=RIGHT)
        b2text = b2.get_tex("x-x_1")

        # Add all objects to the scene
        self.add(line, dot, dot2, b1, b1text, b2, b2text)

```

```python
# Render the scene
render_scene(BraceAnnotation)
```
